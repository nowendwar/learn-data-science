### Movies database

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies
```

```python
movies['title'].sort_values(ascending=False)
```

```python
movies.sort_values('title')
```

```python
filt = (movies['duration'] >= 200) & (movies['genre'] == 'Drama')
```

```python
movies.loc[filt]
```

```python
for m in movies.loc[filt, 'title']:
    print(m)
```

```python
for index, row in movies.loc[filt].iterrows():
    print(index, row.star_rating, row.title)
```
